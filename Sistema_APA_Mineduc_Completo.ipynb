```python
def bloque_regular():
    # Obtener datos
    df_destrezas, mensaje = motor.obtener_destrezas_filtradas()

    # ✅ FUNCIÓN PARA CREAR LA LISTA DESTREZAS (AQUÍ DENTRO, BIEN DEFINIDA)
    def crear_lista_destrezas_multiple(dataframe_destrezas):
        if dataframe_destrezas.empty:
            return widgets.HTML("<p style='color:red; text-align:center;'>❌ NO HAY DESTREZAS PARA ESTA COMBINACIÓN.</p>")
        else:
            opciones = []
            for _, fila in dataframe_destrezas.iterrows():
                icono = ""
                cod_ins = str(fila['insercion_cod'])
                if cod_ins == 'ED': icono = "🌱 "
                elif cod_ins == 'ES': icono = "❤️ "
                elif cod_ins == 'EF': icono = "💰 "
                elif cod_ins == 'EC': icono = "⚖️ "
                elif cod_ins == 'ESV': icono = "🚦 "
                elif cod_ins == 'SEG': icono = "🛡️ "

                etiqueta = f"{icono}[{fila['año']}°] {fila['codigo']} - {fila['descripcion']}"
                opciones.append(etiqueta)

            return widgets.SelectMultiple(
                options=opciones,
                value=[opciones[0]] if opciones else [],
                description='DESTREZAS:',
                layout=Layout(width='95%', height='250px'),
                style={'description_width': 'initial'}
            )

    # ✅ AQUÍ ARMAMOS TODA LA INTERFAZ DEL BLOQUE
    return widgets.VBox([
        widgets.HTML("""
        <div class='tarjeta resaltado-info'>
            <h3 style='color:#154360; margin-top:0;'>📘 BLOQUE: PLANIFICACIÓN REGULAR</h3>
            <p>✅ Metodologías: Incluye ACC/ACI/PIAR | ✅ Selección: Múltiple | ✅ Datos: Completos</p>
        </div>
        """),

        widgets.HTML("<div class='tarjeta'><h4>🧩 METODOLOGÍAS DE ENSEÑANZA</h4>"),
        widgets.SelectMultiple(
            options=metodologias_aprobadas,
            description='Elija:',
            layout=Layout(height='180px', width='90%'),
            value=["Aprendizaje Basado en Problemas"]
        ),
        widgets.HTML("</div>"),

        widgets.HTML("<div class='tarjeta tarjeta-alerta'><h4>🔍 CONTROL DE INSERCIÓN CURRICULAR</h4>"),
        widgets.HTML(f"<p class='resaltado'>{mensaje}</p>"),

        widgets.HTML("<b>📌 SELECCIONE UNA O VARIAS DESTREZAS (Ctrl + Clic):</b>"),
        crear_lista_destrezas_multiple(df_destrezas),

        widgets.HTML("<br><b>❓ ¿LA UNIDAD INVOLUCRA TEMAS DE ENEIS (Ejes Nacionales)?</b>"),
        widgets.RadioButtons(options=['NO', 'SÍ'], description='EJE:', layout=Layout(width='100%')),
        widgets.HTML("</div>")
    ])
    # ✅ INICIO DEL SISTEMA
interfaz_inicio()
```
